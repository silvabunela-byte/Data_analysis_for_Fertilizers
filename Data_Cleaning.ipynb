# Data Cleaning.
This is done so as to ensure that our data are clean so as to ensure thet all the data analysis done are correct.

```python
# Importing the libraries
import pandas as pd
```

```python
# Loading the Merged_data+file.csv
df = pd.read_csv("Data_file/Merged_data_file.csv")
```

```python
# Opening the data frame(df). Here we use the Data Wrangler Extension found in Vs Code.
df
```

```python
"""
Cell generated by Data Wrangler.
"""
def clean_data(df):
    # Replace missing values with 0 in columns: 'month', 'day' and 3 other columns
    df = df.fillna({'month': 0, 'day': 0, 'hour': 0, 'minute': 0, 'second': 0})
    # Replace missing values with 0 in columns: 'month', 'day' and 3 other columns
    df = df.fillna({'month': 0, 'day': 0, 'hour': 0, 'minute': 0, 'second': 0})
    # Drop duplicate rows across all columns
    df = df.drop_duplicates()
    return df

df_clean = clean_data(df.copy())
df_clean.head()
df_clean.to_csv("Cleaned_data.csv", index= False)
```
