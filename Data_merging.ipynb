```python
# The Loading of the Pandas and the glob libraries for the data Loading and Merging.
import pandas as pd
import glob
```

```python
# The data files merging, using the glob Library.
df = glob.glob("Data_file/*.csv")
```

```python
# The file merging, using the pandas Library.
df = pd.concat((pd.read_csv(file) for file in df), ignore_index = True)
```
