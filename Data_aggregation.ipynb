# Data Aggregation.
In the data aggregation we are going to use the cleaned data file for the Descriptive statistics.
It means here we are going to calculate for:
* Mean
* Maximum value
* Minimum value
* Mode
* Count(The number of the items present in a particular column).

```python
# Importing the Pandas Library which will be used for the Loading of our cleaned data file in the/
# data Aggregation.
import pandas as pd
```

```python
# Loading of our data file( Cleaned data file)
df = pd.read_csv("Data_file/Cleaned_data.csv")
```

```python
# Reading of the data file in the Data Frame mode.
df
```

```python
# Loading of the cleaned data file so that the data Aggregation may be done.
     # In the data Aggregation we are going to use the groupby() function for easy analysis/
     # of our data.
     
"""
Cell generated by Data Wrangler.
"""
def clean_data(df):
    # Performed 2 aggregations grouped on columns: 'moisture1', 'moisture2' and 3 other columns
    df = df.groupby(['moisture1', 'moisture2', 'moisture3', 'moisture0', 'second']).agg(day_mean=('day', 'mean'), second_mean=('second', 'mean')).reset_index()
    return df

df_clean = clean_data(df.copy())
df_clean.head()
```
